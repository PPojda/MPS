# Advection-Diffusion Model
This script simulates the concentration c(x, t) of a substance dispersing over time and space,
governed by the advection-diffusion equation:

$$ \frac{\partial c}{\partial t} + U \frac{\partial c}{\partial x} = D \frac{\partial^2 c}{\partial x^2} $$

The fundamental solution for a point source is given by:

$$ c(x,t) = \frac{M}{A} \frac{1}{2 \sqrt{\pi D}} \frac{x}{ t^{3/2}} \exp \left[ -\frac{(Ut - x)^2}{4Dt} \right] $$

where:

* M: Total amount of substance (kg) 
* A: Velocity flux (m³/s)
* D: Diffusion coefficient (m²/s)
* u: Flow velocity (m/s)

```python
import pint
import math
from matplotlib import pyplot
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
si = pint.UnitRegistry() 
```

```python
M = 10*si.kg
A = 100*si.m**3/si.s
u = 0.1*si.m/si.s
D = 10**(-2)*si.m**2/si.s
x = 5*si.m
```

```python
def c(t):
    return (M / A) * (1 / (2 * (np.pi * D)**0.5)) * (x / t**(3/2)) * np.exp(-((u * t - x)**2) / (4 * D * t))
```

```python
si.setup_matplotlib()
t = np.linspace(0, 100, 100)*si.s
pyplot.suptitle("Concentration Distribution Over Time")
pyplot.plot(t, c(t))
```

The plot above represents the point-source distribution of concentration over time for parameters given above. \
The resulting profile follows a Gaussian shape.

Axes:

* x: Time in seconds (s).
* t:  Concentration in kilograms per cubic meter (kg/m³).
